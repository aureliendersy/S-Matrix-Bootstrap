### Plotters for the class 2 amplitude roots associated with $L=2,3$ finite partial wave solutions

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

##### $L=2$

```python
import csv

# Load the correct CSV files with the (L=2) zeros

file_l2z1 = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L2/z1.csv", 'r')
datal2z1 = list(csv.reader(file_l2z1, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
file_l2z2 = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L2/z2.csv", 'r')
datal2z2 = list(csv.reader(file_l2z2, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
```

```python
l2z1re, l2z1im = zip(*datal2z1[::23])
l2z2re, l2z2im = zip(*datal2z2[::23])
```

```python
fig = plt.figure(figsize=(7, 7))

plt.scatter(l2z1re,l2z1im, s=2)
plt.scatter(l2z2re,l2z2im, s=2)
legend = plt.legend(['$z_1$ roots', '$z_2$ roots'], fontsize=12)
for i in range(len(legend.legendHandles)):
    legend.legendHandles[i]._sizes = [10]
tcircle = np.linspace(0,2*math.pi,1000)
plt.plot(np.cos(tcircle), np.sin(tcircle), color='k')
plt.xlabel('Real part', fontsize=14)
plt.ylabel('Imaginary part', fontsize=14)
plt.show()
```

##### $L=3$ zeros

```python
import csv

# Load the correct CSV files with the (L=3) zeros

file_l3z1a = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z1a.csv", 'r')
datal3z1a = list(csv.reader(file_l3z1a, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
file_l3z2a = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z2a.csv", 'r')
datal3z2a = list(csv.reader(file_l3z2a, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
file_l3z3a = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z3a.csv", 'r')
datal3z3a = list(csv.reader(file_l3z3a, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))

file_l3z1b = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z1b.csv", 'r')
datal3z1b = list(csv.reader(file_l3z1b, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
file_l3z2b = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z2b.csv", 'r')
datal3z2b = list(csv.reader(file_l3z2b, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
file_l3z3b = open("/Users/aurelien/PycharmProjects/JupyterNb/roots_finite_L/L3/z3b.csv", 'r')
datal3z3b = list(csv.reader(file_l3z3b, delimiter=',', quoting = csv.QUOTE_NONNUMERIC))
```

```python
l3z1are, l3z1aim = zip(*datal3z1a[4:-1][::3])
l3z2are, l3z2aim = zip(*datal3z2a[4:-1][::3])
l3z3are, l3z3aim = zip(*datal3z3a[4:-1][::3])

l3z1bre, l3z1bim = zip(*datal3z1b[::3])
l3z2bre, l3z2bim = zip(*datal3z2b[::3])
l3z3bre, l3z3bim = zip(*datal3z3b[::3])
```

```python
l3z2arefig2 = np.array(l3z2are)[np.array(l3z2are)<-1]
l3z2arefig1 = np.array(l3z2are)[np.array(l3z2are)>-1]

l3z2aimfig2 = np.array(l3z2aim)[np.array(l3z2are)<-1]
l3z2aimfig1 = np.array(l3z2aim)[np.array(l3z2are)>-1]
```

```python
fig = plt.figure(figsize=(7, 7))

plt.scatter(l3z1are,l3z1aim, s=2)
plt.scatter(l3z2arefig1,l3z2aimfig1, s=2)
plt.scatter(l3z3are,l3z3aim, s=2)


plt.scatter(l3z1bre,l3z1bim, s=2)
plt.scatter(l3z2bre,l3z2bim, s=2)
plt.scatter(l3z3bre,l3z3bim, s=2)

legend = plt.legend(['$z_1$ roots: class 2', '$z_2$ roots: class 2', '$z_3$ roots: class 2', '$z_1$ roots: class 3', '$z_2$ roots: class 3', '$z_3$ roots: class 3'], fontsize=12)
for i in range(len(legend.legendHandles)):
    legend.legendHandles[i]._sizes = [10]
tcircle = np.linspace(0,2*math.pi,1000)
plt.plot(np.cos(tcircle), np.sin(tcircle), color='k')
plt.xlabel('Real part', fontsize=14)
plt.xlim(-2.1,1.1)
plt.ylabel('Imaginary part', fontsize=14)

sub_axes = plt.axes([.2, .15, .2, .2]) 
sub_axes.scatter(l3z2arefig2, l3z2aimfig2, c = 'tab:orange', s=2) 


plt.show()
```
